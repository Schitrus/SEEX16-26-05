```python
import numpy as np
import scipy as sci
import astropy as aspy
import pandas as pd
import matplotlib.pyplot as plt
import astropy.units as u
import jplephem # Need to install
from astropy.coordinates import solar_system_ephemeris
import matplotlib.gridspec as gridspec
from astropy.modeling.models import Trapezoid1D
from matplotlib.gridspec import GridSpec

x=np.linspace(-5,5,1000)
fig = plt.figure(figsize=(12, 8))

def gauss(x):
    x0=0
    res = 0.5*np.e**(-0.5*(x-x0)**2)
    return res


def moffat(x):
    x0=0
    gamma=1/0.4
    alpha=2
    res=0.5* (1+((x-x0)**2)/gamma)**(-alpha)
    return res


def lorentz(x):
    x0=0
    gamma=1/0.4
    res= 1*gamma**2/(gamma**2+(x-x0)**2)
    return res


def ricker(x):
    x0=0
    sigma=1
    res=0.5*(1-((x-x0)**2)/(sigma**2))*np.e**((-(x-x0)**2)/(2*sigma**2))
    return res

s1 = Trapezoid1D()
s1.amplitude = 3
s1.width = 2

gs=fig.add_gridspec(2,6)
ax1= fig.add_subplot(gs[0, 0:2])
ax2= fig.add_subplot(gs[0, 2:4])
ax3= fig.add_subplot(gs[0, 4:6])
ax4= fig.add_subplot(gs[1, 1:3])
ax5= fig.add_subplot(gs[1, 3:5])
ax1.plot(x,gauss(x), lw=2)
ax2.plot(x,moffat(x), lw=2)
ax3.plot(x,lorentz(x), lw=2)
ax4.plot(x,ricker(x), lw=2)
ax5.plot(x, s1(x), lw=2)
ax1.set_xticks([]) 
ax1.set_yticks([])
ax2.set_xticks([]) 
ax2.set_yticks([])
ax3.set_xticks([]) 
ax3.set_yticks([])
ax4.set_xticks([]) 
ax4.set_yticks([])
ax5.set_xticks([]) 
ax5.set_yticks([])
ax1.set_title("Gaussian2D")
ax2.set_title("Moffat2D")
ax3.set_title("Lorentz2D")
ax4.set_title("RickerWavelet2D")
ax5.set_title("TrapezoidDisk2D")
plt.show
```
